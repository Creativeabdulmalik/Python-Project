## TAXI_ORGANIZATION DATASET VISUALIZATION

###  Bar chart below visualize the trip_purpose count within the 3 months of operation

```python
# total Count the occurrences of each category
category_counts = df['trip_purpose'].value_counts()

# bar chart
category_counts.plot(kind='bar', color='darkblue',   edgecolor='black', title='TRIP_PURPOSE COUNT')
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

### Bar chart count the total traffic  between the three variables of light,medium and heavy traffic

```python
# total Count the occurrences of each category
category_counts = df['traffic'].value_counts()

# bar chart
category_counts.plot(kind='bar', color='purple',   edgecolor='black', title='traffic Counts')
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

### Plot a histogram

```python
df['total_amount'].plot(kind='hist', bins=10, color='lightgreen', edgecolor='black', title='Tip_Amount')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

### The scatter chart visualize the relationship between fare amount paid by each customers and trip distance covered during the trip

```python
# Scatter plot
plt.scatter(df['fare_amount'], df['trip_distance'], color='purple', alpha=0.7)  
plt.title("Fare_Amount by Trip_Distance")
plt.xlabel("fare_amount")
plt.ylabel("trip_distance")
plt.show()
```

### The box plot visualize the relationship between payment method type by total revenue generated

```python
# Box plot
sns.boxplot(data=df, x='payment_type', color='red', y='total_amount')
plt.title("Payment type by Total Revenue generated")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# taxi_organization dataset
taxi_organization = "taxi_organization.csv"
df = pd.read_csv(taxi_organization)

# Aggregate data of location
category_counts = df['location'].value_counts()

# Pie Chart of location
category_counts.plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(8, 8),
    startangle=90,
    colors=plt.cm.Paired.colors
)
plt.title("Pie Chart")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# taxi_organization the dataset
file_path = "taxi_organization.csv"
df = pd.read_csv(file_path)

df['pickup_datetime'] = pd.to_datetime(df['dropoff_datetime'])
df = df.sort_values(by='pickup_datetime')

# Plot the line graph
plt.figure(figsize=(10, 6))
plt.plot(df['pickup_datetime'], df['trip_distance'], marker='o', linestyle='-', color='red', label='Trend')

# Add labels and title
plt.title("Trend and Pattern Over Time", fontsize=16)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Values", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.savefig()
plt.show()
```

### The scatter plot chart Visualize the relationship between weather and trip purpose

```python
# Scatter plot
plt.scatter(df['weather'], df['trip_purpose'], color='darkgreen', alpha=0.7)  
plt.title("Scatter Plot")
plt.xlabel("weather")
plt.ylabel("trip_purpose")
plt.show()
```

### Box plot is showing us the relationship between weather and hours of day

```python
# Box plot
import seaborn as sns

sns.boxplot(data=df, x='weather', color='darkgreen', y='hour_of_day')
plt.title("weather_by_ hours of day")
plt.show()
```
